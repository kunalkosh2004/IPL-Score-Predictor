```python
import pandas as pd
import numpy as np
import pickle
```

```python
df = pickle.load(open('data1.pkl','rb'))
```

```python
df.iloc[100]
```

```python
df.columns
```

```python
df.drop(columns=['meta.data_version','meta.created','meta.revision','info.dates','info.player_of_match','info.toss.decision','info.toss.winner','info.umpires','info.outcome.by.wickets','info.outcome.method',
       'info.neutral_venue', 'info.outcome.eliminator', 'info.outcome.result'],inplace=True)
```

```python
df.drop(columns=['info.overs','info.match_type','info.gender'],inplace=True)
```

```python
df.iloc[0]['innings'][0]['1st innings']
```

```python
count=1
df2 = pd.DataFrame()

for index,row in df.iterrows():
    count+=1
    ball_no=[]
    batsman=[]
    bowler=[]
    teams=[]
    total_run=[]
    player_dismissed=[]
    batting_team=[]
    match_id=[]
    city=[]
    venue=[]
    for ball in row['innings'][0]['1st innings']['deliveries']:
        for i in ball:
            match_id.append(count)
            ball_no.append(i)
            batsman.append(ball[i]['batsman'])
            bowler.append(ball[i]['bowler'])
            teams.append(row['info.teams'])
            total_run.append(ball[i]['runs']['total'])
            try:
                player_dismissed.append(ball[i]['wicket']['player_out'])
            except:
                player_dismissed.append(0)
            batting_team.append(row['innings'][0]['1st innings']['team'])
            city.append(row['info.city'])
            venue.append(row['info.venue'])
    loop_df = pd.DataFrame({
            'match_id':match_id,
            'teams':teams,
            'batting_team':batting_team,
            'ball':ball_no,
            'batsman':batsman,
            'bowler':bowler,
            'runs':total_run,
            'player_dismissed':player_dismissed,
            'city':city,
            'venue':venue
        })
    df2 = pd.concat([df2,loop_df],ignore_index=True)
```

```python
df2['city'].fillna(df2['venue'].apply(lambda x: x.split()[0]),inplace=True)
```

```python
def get_bowling_team(row):
    for i in row['teams']:
        if i!=row['batting_team']:
            return i
```

```python
df2['bowling_team'] = df2.apply(get_bowling_team,axis=1)
```

```python
df2.drop(columns=['venue','teams'],inplace=True)
```
