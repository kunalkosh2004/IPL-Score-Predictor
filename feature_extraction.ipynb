```python
import pickle
```

```python
df = pickle.load(open('data2.pkl','rb'))
```

```python
df
```

```python
df['batting_team']=df['batting_team'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
df['batting_team']=df['batting_team'].str.replace('Rising Pune Supergiants','Pune Warriors')
df['batting_team']=df['batting_team'].str.replace('Rising Pune Supergiant','Pune Warriors')
df['batting_team']=df['batting_team'].str.replace('Delhi Daredevils','Delhi Capitals')
df['batting_team']=df['batting_team'].str.replace('Rising Pune Supergiants','Pune Warriors')
df['batting_team']=df['batting_team'].str.replace('Rising Pune Supergiants','Pune Warriors')
```

```python
df['bowling_team']=df['bowling_team'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
df['bowling_team']=df['bowling_team'].str.replace('Rising Pune Supergiants','Pune Warriors')
df['bowling_team']=df['bowling_team'].str.replace('Rising Pune Supergiant','Pune Warriors')
df['bowling_team']=df['bowling_team'].str.replace('Delhi Daredevils','Delhi Capitals')
df['bowling_team']=df['bowling_team'].str.replace('Rising Pune Supergiants','Pune Warriors')
df['bowling_team']=df['bowling_team'].str.replace('Rising Pune Supergiants','Pune Warriors')
```

```python
df=df[df['batting_team']!='Kochi Tuskers Kerala']
```

```python
df['city'].value_counts().index.tolist()
```

```python
df['over_no']=df['ball'].astype('str').apply(lambda x: x.split('.')[0])
df['ball_no']=df['ball'].astype('str').apply(lambda x: x.split('.')[1])
```

```python
city_list=df['city'].value_counts()[df['city'].value_counts()>800].index.tolist()
```

```python
df=df[df['city'].isin(city_list)]
```

```python
df
```

```python
df['total_ball']=df['over_no'].astype('int32')*6+df['ball_no'].astype('int32')
```

```python
df['balls_left'] = 120-df['total_ball']
```

```python
df['balls_left'] = df['balls_left'].apply(lambda x: x if x>0 else 0)
```

```python
df['curr_score']=df.groupby('match_id')['runs'].cumsum()
```

```python
df['player_dismissed']=df['player_dismissed'].apply(lambda x: 0 if x==0 else 1)
```

```python
df['wickets_gone'] = df.groupby('match_id')['player_dismissed'].cumsum()
```

```python
df['wickets_left'] = 10-df['wickets_gone']
```

```python
df['ball_no']=df['ball_no'].astype('int32')
```

```python
df['crr'] = (df['curr_score']*6)/df['total_ball']
```

```python
groups = df.groupby('match_id')

match_ids = df['match_id'].unique()
last_five=[]
for i in match_ids:
    last_five.extend(groups.get_group(i).rolling(window=30)['runs'].sum().values.tolist())
```

```python
df['last_five_runs'] = last_five
```

```python
final_df = df.groupby('match_id')['runs'].sum().reset_index().merge(df,on='match_id')
```

```python
final_df=final_df[['batting_team','bowling_team','city','curr_score','balls_left','wickets_left','crr','last_five_runs','runs_x']]
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.dropna(inplace=True)
```

```python
pickle.dump(final_df,open('final_data.pkl','wb'))
```
